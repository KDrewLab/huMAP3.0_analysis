```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
save_path = 'Uniplex_20230503_intact_subset10kNEG_notScaled_accuracy/'
```

```python
df = TabularDataset("humap2_15k_bioplex293T_HCT116_withWMM_reviewedACC_20220625aa.featmat")
```

```python
df_trim = df.drop(["acc1","acc2"], axis=1)
```

```python
df_trim.head()
```

```python
predictor = TabularPredictor.load(save_path)
y_pred = predictor.predict(df_trim)
print("Predictions:  \n", y_pred)
```

```python
pred_df_probs = predictor.predict_proba(df_trim)
```

```python
pred_df_probs.sort_values(1)
```

```python
pred_df_probs[1]
```

```python
df['pred_prob'] = pred_df_probs[1]
```

```python
df.head()
```

```python
df.sort_values("pred_prob", ascending=False)[['acc1','acc2','pred_prob']].to_csv("ComplexPortal_model_unlabeledpredictions_aa_20230523.pairsWprob", index=False, header=False, sep='\t')
```
