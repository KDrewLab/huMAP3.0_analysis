```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
train = TabularDataset("./humap3_train_20220625.featmat")
```

```python
train_trim = train.drop(["acc1","acc2"], axis=1)
```

```python
label = 'ComplexPortal_train'
```

```python
save_path = './huMAP3_20230503_complexportal_subset10kNEG_notScaled_accuracy/'  # specifies folder to store trained models
predictor = TabularPredictor(label=label, eval_metric="accuracy", path=save_path).fit(train_trim)
```

```python
test = TabularDataset("./humap3_test_20220625.featmat")
```

```python
test_trim = test.drop(["acc1","acc2"], axis=1)
```

```python
test_label = 'ComplexPortal_test'
y_test = test_trim[test_label]  # values to predict
test_nolab = test_trim.drop(columns=[test_label])  # delete label column 
test_nolab.head()
```

```python
predictor = TabularPredictor.load(save_path)
y_pred = predictor.predict(test_nolab)
print("Predictions:  \n", y_pred)
perf = predictor.evaluate_predictions(y_true=y_test, y_pred=y_pred, auxiliary_metrics=True)
```

```python
#kdrew: autogluon expects label to be "train" when predicting
test_trim = test_trim.rename(columns={"ComplexPortal_test":"ComplexPortal_train"})
```

```python
predictor.leaderboard(test_trim, silent=True)
```

```python
pred_test_probs = predictor.predict_proba(test_nolab)
```

```python
pred_test_probs.sort_values(1)
```

```python
test['pred_prob'] = pred_test_probs[1]
```

```python
test.sort_values("pred_prob",ascending=False)[['acc1','acc2','IntAct_test','pred_prob']].head(70)
```
