```python
#sfisch: we import autogluon to do the training for our model - Version 0.4.0 is needed to use with our model
from autogluon.tabular import TabularPredictor
#sfisch6: we pull the training feature matrix from huggingface using the datasets module
from datasets import load_dataset
```

```python
#sfisch6: load train data from huggingface
dataset = load_dataset('sfisch/hu.MAP3.0')
df_train = dataset["train"].to_pandas()
```

```python
df_train.head()
```

```python
#sfisch6: we drop the protein identifiers and train with just the features and the labels 
train_trim = df_train.drop(["acc1","acc2"], axis=1)
```

```python
#sfisch6: define label column to pass to model for training - note the model will expect the same label column
# in training and testing
label = 'label'
```

```python
#sfisch6: define folder to store trained models
save_path = './huMAP3_20230503_complexportal_subset10kNEG_notScaled_accuracy/' 
```

```python
#sfisch6: train model by passing training feature matrix, label col, and path for saved models
predictor = TabularPredictor(label=label, eval_metric="accuracy", path=save_path).fit(train_trim)
```
